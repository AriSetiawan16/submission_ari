## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Pastikan datetime dalam format yang benar
if 'datetime' not in df.columns:
    df.reset_index(inplace=True)

df['datetime'] = pd.to_datetime(df['datetime'])

# Kelompokkan data per bulan untuk melihat tren perubahan kualitas udara
df['year_month'] = df['datetime'].dt.to_period('M')

# Hitung rata-rata bulanan untuk PM2.5
monthly_pm25 = df.groupby('year_month')['PM2.5'].mean()

# Plot tren perubahan PM2.5
plt.figure(figsize=(12, 6))
monthly_pm25.plot(marker='o', linestyle='-')
plt.title("Tren Perubahan PM2.5 di Beijing (2013-2017)")
plt.xlabel("Waktu (Tahun-Bulan)")
plt.ylabel("Konsentrasi PM2.5 (g/m)")
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Cari periode dengan tingkat pencemaran tertinggi
highest_pollution_period = monthly_pm25.idxmax()
highest_pm25_value = monthly_pm25.max()

print(f"Periode dengan tingkat pencemaran tertinggi: {highest_pollution_period}")
print(f"Rata-rata PM2.5 tertinggi: {highest_pm25_value:.2f} g/m")
```
